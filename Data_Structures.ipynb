List

```python
L =[1,2,3,4,"asim"]
print(L)
print(L[0])
print(L[1:4])
L = L + ["how", " are", 4]
print(L)
L.append(7)
print(L)
```

Tuple

```python
T = (1,2,3,4,"asim")
print(T)
print(T[1])
print(T[:5])
T2 = (4,2,5,"tUPLE KHAN")
T3 = T + T2
print(T3)
```

Set

```python
S = {1,2,3,4,"asim"}
print(S)
print(3 in S)
```

Dictionary

```python
D = {"Banana" : 12, "Asim" : 10, }
print(D)
print(D["Asim"])
D[1] = "DIC"
print(D)
D2 = {
    "name": "Ali",
    "age": 20,
    "course": "Python"
}
D.update(D2)
print(D2)
```

```python
def getdata():
    D = {}
    while True:
        stdid = int(input("Enter student id: "))
        marks = input("Enter student marks separated by comma: ")
        morestudents = input("Enter 'no' to quit: ")

        if stdid in D:
            print(stdid, "Already exists!")
        else:
            D[stdid] = marks.split(",")

        if morestudents.lower() == "no":
            return D

student_data = getdata()
```
