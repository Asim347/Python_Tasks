Given a list of numbers L = [1,2,3,4,5,-1,9,2] and make another list that contains all items in ascending order from min to max.

```python
L = [1,2,3,4,5,-1,9,2]

for i in range(len(L)):
    min_index = i
    for j in range(i+1, len(L)):
        if L[j] < L[min_index]:
            min_index = j

    # swap
    temp = L[i]
    L[i] = L[min_index]
    L[min_index] = temp

print(L)
```
